```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
# CSV: Comma Separated Value, 콤마로 구분된 값이라는 뜻
# Pandas에 csv를 읽어드리는 함수가 있음, sep: 구분자
abalone_data = pd.read_csv(os.path.join('data','abalone.csv'), 
                 sep=',', 
                 names=['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings'],
                 header=None
)

abalone_data.shape
```

```python
abalone_data.head()
```

```python
label = abalone_data['Sex'] # labeling
del abalone_data['Sex'] # "Sex" column 삭제

abalone_data.shape
```

```python
abalone_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler # 변수 정규화

scaler = MinMaxScaler()
scaler.fit(abalone_data)
norm_data = scaler.fit_transform(abalone_data)
print(norm_data)
print('최대: ',np.max(norm_data),', 최소: ',np.min(norm_data))
```

```python
from sklearn.preprocessing import LabelBinarizer

encoder = LabelBinarizer()
transformed_label = encoder.fit_transform(label)

print(encoder.classes_)
print(transformed_label)
```

```python
# Sklearn의 train_test_split 함수를 써서 train/test data 분할
from sklearn.model_selection import train_test_split

trnx, tstx, trny, tsty = train_test_split(norm_data, transformed_label, test_size=0.3, random_state=111)
```

```python
# Sklearn을 이용해서 Classifier A를 만든다(예제와 다른 옵션 사용)
from sklearn.neural_network import MLPClassifier

clf_A = MLPClassifier(hidden_layer_sizes=(100,200,400,200,100,))
clf_A.fit(trnx, trny)
tsty_hat = clf_A.predict(tstx)
```

```python
print(clf_A) # cclassifier A의 정보 확인
```

```python
# 값이 얼마나 맞았나 대략적으로 확인
print(tsty[0:20])
print(tsty_hat[0:20])
```

```python
from sklearn.metrics import accuracy_score
accuracy_A = accuracy_score(tsty, tsty_hat)
print(accuracy_A)
```

```python
# keras를 이용해서 Classifier B를 만든다(예제와 다른 옵션 사용)
import tensorflow as tf
import keras
from keras import layers, models, optimizers

input_shape = (8,)

mlp_model = models.Sequential()
mlp_model.add(layers.Dense(units = 100, activation = 'relu', input_shape=input_shape))
mlp_model.add(layers.Dense(units = 200, activation = 'relu'))
mlp_model.add(layers.Dense(units = 100, activation = 'relu'))
mlp_model.add(layers.Dense(units = 3, activation = 'softmax'))

mlp_model.compile(optimizer='Adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
mlp_model.summary()
```

```python
history = mlp_model.fit(trnx, trny, validation_data = [tstx, tsty], batch_size=100, epochs=35)
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
accuracy_B = history.history['val_acc'][-1]
print(accuracy_B)
```
